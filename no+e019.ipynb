```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.size'] = 14 # デフォルトの文字サイズ
w = 1280; h = 670; r = w/h; Size = 5.0
fig, ax = plt.subplots(figsize=(r*Size,Size))

W = np.pi # x軸の表示範囲
ax.set_xlim(-W,W)
ax.set_ylim(-W/r, W/r)
ax.set_xticks(np.linspace(-W, W, 5))
ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'],
                    math_fontfamily='cm')
ax.set_yticks(np.linspace(-1, 1, 3))
ax.set_yticklabels([r'$-1$', r'$0$', r'$1$'],
                    math_fontfamily='cm')
ax.grid()

x = np.linspace(-np.pi, np.pi, 2**10 + 1)
y = np.cos(x)

N = 20
for k in range(1,N):
  y = y + np.cos(2**k * x) / 2**k

ax.plot(x,0.5*y, linewidth=0.5)
plt.savefig("math019.png",format="png",dpi=300)
plt.show()
```
