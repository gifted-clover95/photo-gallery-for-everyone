```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def phi_internal(x):
  if (0.0<=x) & (x<=0.5):
    retv = 2.0*x
  elif (0.5<x) & (x<=1.0):
    retv = 2.0*(1.0-x)
  else:
    retv = 0.0
  return retv
```

```python
phi = np.frompyfunc(phi_internal, 1, 1)
```

```python
plt.rcParams['font.size'] = 14 # デフォルトの文字サイズ
r = 5.0
```

```python
fig, ax = plt.subplots(figsize=(1.91*r,r))

N = 10
for k in range(1,N):
  x = np.linspace(0.0, 1.0, 2**k + 1)
  y = phi(x)
  for i in range(k-1):
    y = phi(y)
  ax.plot(x,y * 2**(-k))

plt.show()
```
