今回は、[Wikipedia の Vicsek fractal](https://en.wikipedia.org/wiki/Vicsek_fractal) で紹介されている図を Python で描いていきます。この記事で作成した画像も、みんなのフォトギャラリーに登録する予定です。

この記事は、細かい理屈は無視して画像を目で見て楽しんで頂くつもりで書いているので、図の作成方法の説明は省きます。

```python
# ライブラリのインポート
import matplotlib.pyplot as plt
import numpy as np
```

```python
def zoom_in (img, n):
    s1, s2 = img.shape
    retv = np.zeros((s1*n,s2*n))
    block = np.ones((n,n))
    for i in range(s1):
        for j in range(s2):
            retv[i*n:(i+1)*n, j*n:(j+1)*n] = img[i][j] * block
    return retv
```

```python
img_count = 0 # 画像の枚数
N = 5; r = 3 ** N
a = 3
im = np.ones(a*a).reshape(a,a)
im[0][1] = 0
im[1][0] = 0
im[1][1] = 0
im[1][2] = 0
im[2][1] = 0

plt.imshow(im, cmap='gray')
plt.axis('off')
plt.show()

im2 = zoom_in (im, r)
img_count += 1
plt.imsave('no+e_photo_gallery_029_'+str(img_count).zfill(3)+'.png', im2, cmap='gray')
print(a)

for i in range(N):
    prev = im
    prev_a = a

    a = prev_a * 3
    im = np.ones(a*a).reshape(a,a)
    im[:prev_a, prev_a:prev_a*2] = prev
    im[prev_a:prev_a*2, :prev_a] = prev
    im[prev_a:prev_a*2, prev_a:prev_a*2] = prev
    im[prev_a:prev_a*2, prev_a*2:] = prev
    im[prev_a*2:, prev_a:prev_a*2] = prev

    plt.imshow(im, cmap='gray')
    plt.axis('off')
    plt.show()

    r = r // 3
    im2 = zoom_in(im,r)
    img_count += 1
    plt.imsave('no+e_photo_gallery_029_'+str(img_count).zfill(3)+'.png', im2, cmap='gray')
    print(i, a, im2.shape)
```

```python
# "みんなのフォトギャラリー"の画像サイズ
width = 1280 ; height = 670
```

```python
# サイズ調整
width = width * 11 // 10
height = height * 11 // 10
```

```python
h0 , w0 = im.shape
h = (height-h0)//2
w = (width-w0)//2
img = np.ones(height*width).reshape(height,width)
img[h:(h+h0), w:(w+w0)] = im
```

```python
#画像の表示と保存
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()

img_count += 1
plt.imsave('no+e_photo_gallery_029_'+str(img_count).zfill(3)+'.png', img, cmap='gray')
```

```python
# "みんなのフォトギャラリー"の画像サイズ
width = 1280 ; height = 670
```

```python
a = 3
im = np.ones(a*a).reshape(a,a)
im[0][0] = 0
im[0][2] = 0
im[1][1] = 0
im[2][0] = 0
im[2][2] = 0
for i in range(3):
    prev = im
    prev_a = a

    a = prev_a * 3 ; N += 1
    im = np.ones(a*a).reshape(a,a)
    im[:prev_a, :prev_a] = prev
    im[:prev_a, prev_a*2:] = prev
    im[prev_a:prev_a*2, prev_a:prev_a*2] = prev
    im[prev_a*2:, :prev_a] = prev
    im[prev_a*2:, prev_a*2:] = prev

plt.imshow(im, cmap='gray')
plt.axis('off')
plt.show()
print(i, a)
```

```python
h, w = im.shape
img = np.ones(height*width).reshape(height, width)
h1 = 0; h2 = h
while h1 < height:
    w1 = 0; w2 = w;
    while w1 < width:
        if h2 < height:
            if w2 < width:
                img[h1:h2,w1:w2] = im
            else:
                img[h1:h2,w1:] = im[:,:(width-w2)]
        else:
            if w2 < width:
                img[h1:,w1:w2] = im[:(height-h2),:]
            else:
                img[h1:,w1:] = im[:(height-h2),:(width-w2)]
        w1 = w2; w2 += w
    h1 = h2; h2 += h

plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
img_count += 1
plt.imsave('no+e_photo_gallery_029_'+str(img_count).zfill(3)+'.png', img, cmap='gray')
```
