```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class photo_gallery_for_everyone :
    def __init__(self, file_name):
        self.file_name = file_name
        self.count = 0
        self.height = 670
        self.width = 1280
        self.img = np.ones((self.height, self.width, 3), dtype=np.uint8) * 255
        self.origin = (self.height, self.width)
        self.scale = 1.0

    def set_origin_center(self):
        self.origin = (self.height//2, self.width//2)

    def show(self):
        #plt.imshow(self.img)
        #plt.show()
        self.count += 1
        plt.imsave(self.file_name + str(self.count).zfill(3) + '.png', self.img)

    def show2(self, img):
        #plt.imshow(img)
        #plt.show()
        self.count += 1
        plt.imsave(self.file_name + str(self.count).zfill(3) + '.png', img)
```

```python
photo_gallery = photo_gallery_for_everyone('no+e_photo_gallery_034_')
```

```python
class color_iterator():
    def __init__(self, initial = 0):
        self.edge, self.r =  divmod (initial, 256)

    def get_color(self):
        if self.edge == 0:
            retv = (255, 255-self.r, self.r)
        elif self.edge == 1:
            retv = (255-self.r, self.r, 255)
        else:
            retv = (self.r, 255, 255-self.r)

        self.r += 1
        if self.r == 256:
            self.r = 1
            self.edge += 1
            if self.edge == 3:
                self.edge = 0

        return retv
```

```python
class color_iterator2():
    def __init__(self, initial = 0):
        self.edge, self.r =  divmod (initial, 256)

    def get_color(self):
        if self.edge == 0:
            retv = (255-self.r, self.r, 0)
        elif self.edge == 1:
            retv = (0, 255-self.r, self.r)
        else:
            retv = (self.r, 0, 255-self.r)

        self.r += 1
        if self.r == 256:
            self.r = 1
            self.edge += 1
            if self.edge == 3:
                self.edge = 0

        return retv
```

```python
class curve:
    def __init__(self, n, prm1 = 2, prm2 =0):
        self.n = n
        self.prm1 = prm1
        self.prm2 = prm2
        self.str = self.gen_str(n)
        self.points = self.gen_points (n)
        self.color = (0,0,0)
        self.x0 = 0
        self.y0 = 0
        self.direction = 0

    # 文字列を生成するための関数
    def gen_str(self, n):
        s = 'A'
        for i in range(n):
            s = s.translate(str.maketrans({'A':'+BF-AFA-FB+','B':'-AF+BFB+FA-'}))
        return s

    # 亀の向きを変える
    def L(self,l,d0):
        d = (d0+1) % 4
        for i in range(self.prm2):
            self.F(l,d)
            self.F(l,d0)
        return d
    def R(self,l,d0):
        d = (d0-1) % 4
        for i in range(self.prm2):
            self.F(l,d)
            self.F(l,d0)
        return d
    # 亀を前進させる
    def F(self,l, direction):
        x = l[-1][0]; y = l[-1][1]
        if direction == 0:
            l.append([x+1, y  ])
        elif direction == 1:
            l.append([x  , y+1])
        elif direction == 2:
            l.append([x-1, y  ])
        elif direction == 3:
            l.append([x  , y-1])
        else:
            print("F:error")

    # 生成された文字列にしたがって亀を移動させる
    def move(self,l,c,direction):
        if c == 'F':
            for i in range(self.prm1):
                self.F(l,direction)
            retv = direction
        elif c == '+':
            retv = self.L(l,direction)
        elif c == '-':
            retv = self.R(l,direction)
        else:
            retv = direction
        return retv

    # 点の座標の生成
    def gen_points(self, n):
        l = [[0,0]]
        d = 0
        for c in self.str:
            d = self.move(l,c,d)

        n_points = len(l)
        x = np.zeros((n_points,2),dtype=int)
        for i in range(n_points):
            x[i][0] = l[i][0]
            x[i][1] = l[i][1]

        self.min_x = min(x[:,0])
        self.min_y = min(x[:,1])
        self.max_x = max(x[:,0])
        self.max_y = max(x[:,1])

        self.width = self.max_x - self.min_x+1
        self.height = self.max_y - self.min_y+1
        return x

    def draw (self,im, x, y, direction=0, color=(0,0,0)):
        self.color = color
        self.direction = direction
        self.x0 = x
        self.y0 = y
        if self.direction == 0:
            for pt in self.points:
                im[self.x0+pt[0], self.y0+pt[1],0] = self.color[0]
                im[self.x0+pt[0], self.y0+pt[1],1] = self.color[1]
                im[self.x0+pt[0], self.y0+pt[1],2] = self.color[2]
        elif self.direction == 1:
            for pt in self.points:
                im[self.x0-pt[1], self.y0+pt[0],0] = self.color[0]
                im[self.x0-pt[1], self.y0+pt[0],1] = self.color[1]
                im[self.x0-pt[1], self.y0+pt[0],2] = self.color[2]
        elif self.direction == 2:
            for pt in self.points:
                im[self.x0-pt[0], self.y0-pt[1],0] = self.color[0]
                im[self.x0-pt[0], self.y0-pt[1],1] = self.color[1]
                im[self.x0-pt[0], self.y0-pt[1],2] = self.color[2]
        elif self.direction == 3:
            for pt in self.points:
                im[self.x0+pt[1], self.y0-pt[0],0] = self.color[0]
                im[self.x0+pt[1], self.y0-pt[0],1] = self.color[1]
                im[self.x0+pt[1], self.y0-pt[0],2] = self.color[2]

    def draw2 (self,im, x, y, col_iter, direction=0):
        self.direction = direction
        self.x0 = x
        self.y0 = y
        if self.direction == 0:
            for pt in self.points:
                self.color = col_iter.get_color()
                im[self.x0+pt[0], self.y0+pt[1],0] = self.color[0]
                im[self.x0+pt[0], self.y0+pt[1],1] = self.color[1]
                im[self.x0+pt[0], self.y0+pt[1],2] = self.color[2]
        elif self.direction == 1:
            for pt in self.points:
                self.color = col_iter.get_color()
                im[self.x0-pt[1], self.y0+pt[0],0] = self.color[0]
                im[self.x0-pt[1], self.y0+pt[0],1] = self.color[1]
                im[self.x0-pt[1], self.y0+pt[0],2] = self.color[2]
        elif self.direction == 2:
            for pt in self.points:
                self.color = col_iter.get_color()
                im[self.x0-pt[0], self.y0-pt[1],0] = self.color[0]
                im[self.x0-pt[0], self.y0-pt[1],1] = self.color[1]
                im[self.x0-pt[0], self.y0-pt[1],2] = self.color[2]
        elif self.direction == 3:
            for pt in self.points:
                self.color = col_iter.get_color()
                im[self.x0+pt[1], self.y0-pt[0],0] = self.color[0]
                im[self.x0+pt[1], self.y0-pt[0],1] = self.color[1]
                im[self.x0+pt[1], self.y0-pt[0],2] = self.color[2]
```

```python
def zoom_in (img, n):
    s1, s2, ncol = img.shape
    retv = np.zeros((s1*n, s2*n, ncol), dtype=np.uint8)
    block = np.ones((n,n))
    for i in range(s1):
        for j in range(s2):
            for k in range(ncol):
                retv[i*n:(i+1)*n, j*n:(j+1)*n,k] = img[i][j][k] * block
    return retv
```

```python
for n in range(1,8):
    h_curve = curve(n,3,0)
    c = color_iterator()
    M = max([h_curve.max_x - h_curve.min_x, h_curve.max_y-h_curve.min_y])
    im = np.zeros((M+1, M+1, 3), dtype=np.uint8)
    im[:,:,:] = 50

    h_curve.draw2(im, M, 0, c, 1)

    k = photo_gallery.height // (M+1) + 1
    height = k * (M + 1)
    width = (photo_gallery.width * height) // photo_gallery.height

    im = zoom_in (im, k)
    h, w, c = im.shape
    img = np.zeros((height, width, 3), dtype=np.uint8)
    img[:,:,:] = 50
    dh = (height - h)//2; dw = (width - w)//2
    img[dh:(dh+h), dw:(dw+w),:] = im

    photo_gallery.show2(img)
```
