```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import sieve
```

```python
N = 10**5
p = np.array([i for i in sieve.primerange(2, N)])
```

```python
# 極座標
x = p * np.cos(p)
y = p * np.sin(p)
```

```python
r = 5.0
fig, ax = plt.subplots(figsize=(1.91*r,r))
ax.axis("off")
W = np.max(p) # x軸の表示範囲
ax.set_xlim(-W,W)
ax.set_ylim(-W/1.91, W/1.91)

ax.scatter(x,y, s = 1)
plt.show()
```
