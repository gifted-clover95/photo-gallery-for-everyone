```python
import matplotlib.pyplot as plt
import numpy as np

# 亀の向きを変える
def L(l):
  l[0] = (l[0]-1) % 6
  return

def R(l):
  l[0] = (l[0]+1) % 6
  return

# 亀を前進させる
def F(l):
  direction = l[0]
  if direction == 0:
    l.insert(1,[l[1][0]+1, l[1][1]  ])
  elif direction == 1:
    l.insert(1,[l[1][0]  , l[1][1]+1])
  elif direction == 2:
    l.insert(1,[l[1][0]-1, l[1][1]+1])
  elif direction == 3:
    l.insert(1,[l[1][0]-1, l[1][1]  ])
  elif direction == 4:
    l.insert(1,[l[1][0]  , l[1][1]-1])
  elif direction == 5:
    l.insert(1,[l[1][0]+1, l[1][1]-1])
  else:
    print("F:error")

# 生成された文字列にしたがって亀を移動させる
def move(l,c):
  if c == '-':
    L(l)
  elif c == '+':
    R(l)
  elif c == 'F':
    F(l)
  else:
    print('move:error')

# 文字列を生成するための関数
def generator(c):
  if c == 'F':
    retv = 'F+F--F+F'
  else:
    retv = c
  return retv

def gen_str(s, n):
  for i in range(n):
    #print(s)
    tmp = ''
    for c in s:
      tmp = tmp + generator(c)
    s = tmp
  return s

# パスを生成する関数
def make_path(s):
  l = [ 0, [0,0]]
  for c in s:
    move(l,c)
  # リストを配列に変換
  length = len(l)-1
  A = np.zeros((length, 2))
  for i in range(length):
    A[i,0] = l[length-i][0]
    A[i,1] = l[length-i][1]
  s30 = np.sin(np.pi/3); c30 = np.cos(np.pi/3)
  Base = np.matrix([[1.0, 0.0], [c30, s30]])
  return np.matmul(A, Base)
```

```python
for n in range(5):
    s = gen_str('F--F--F', n) # 文字列の生成
    #print(s)
    X = make_path(s)    # 座標の取得
    #print(X)

    # 図形の描画
    plt.rcParams['font.size'] = 14 # デフォルトの文字サイズ
    w = 1280; h = 670; r = w/h; Size = 5.0
    fig, ax = plt.subplots(figsize=(r*Size,Size))
    ax.axis("off")

    y_max = np.max(X[:,1]); y_min = np.min(X[:,1])
    H = y_max - y_min
    W = r*H
    x_mid = 0.5*(np.max(X[:,0])+np.min(X[:,0]))
    ax.set_xlim(x_mid-0.5*W, x_mid+0.5*W)
    ax.set_ylim(y_min,y_max)

    ax.plot(X[:,0], X[:,1])
    plt.savefig("math021-"+str(n)+".png",format="png",dpi=300)
    plt.show()
```
