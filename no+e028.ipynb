```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# 「みんなのフォトギャラリー」用のサイズ
width = 1280; height = 670
```

```python
# エラトステネスの篩で素数の一覧を返す
def Sieve_of_Eratosthenes (N):
    s = np.zeros(N-1)
    primes = []
    pt = 0
    while pt < N-1:
        if s[pt] == 0:
            val = pt+2
            primes.append (val)
            pt2 = pt + val
            while pt2 < N-1:
                s[pt2] = 1
                pt2 += val
        pt+=1
    return primes, s
```

```python
N = 10000
primes, s = Sieve_of_Eratosthenes (N)
```

```python
n = 10
offset_w = 50
offset_h = 50
w = (width-2*offset_w) // n
h = (height-2*offset_h) // n

img = np.full((height,width,3),255, dtype=np.uint8)

for i in range(n):
    for j in range(n):
        val = str(n*i+j+2).rjust(3,' ')
        if s[n*i+j] == 0:
            col = (0xff,0x8c,0x00)
        else :
            col = (0x0e,0x7f,0xff)
        cv2.putText(img, val,
                    (w * j+offset_w,
                     h *(i+1)+offset_h),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    2.0, col, thickness=2)

plt.imsave('no+e_photo_gallery_028_1.png', img, cmap='gray',dpi=1000 )
plt.imshow( cv2.cvtColor(img, cv2.COLOR_BGR2RGB) )
```
