```python
import matplotlib.pyplot as plt
import numpy as np

def phi_internal(x):
  if (0.0<=x) & (x<=0.5):
    retv = 2.0*x
  elif (0.5<x) & (x<=1.0):
    retv = 2.0*(1.0-x)
  else:
    retv = 0.0
  return retv

phi = np.frompyfunc(phi_internal, 1, 1)

plt.rcParams['font.size'] = 14 # デフォルトの文字サイズ
r = 5.0
fig, ax = plt.subplots(figsize=(1.91*r,r))

N = 20
x = np.linspace(0.0, 1.0, 2**N + 1)
y = x
T = np.zeros(2**N + 1)
for k in range(1,N):
  y = phi(y)
  T = T + y/(2**k)

ax.plot(x,T, linewidth=0.5)
plt.show()
```
