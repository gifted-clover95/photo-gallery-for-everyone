```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['font.size'] = 14 # デフォルトの文字サイズ
r = 5.0
fig, ax = plt.subplots(figsize=(1.91*r,r))
eps = 0.1
W = np.pi # x軸の表示範囲
ax.set_xlim(-W,W)
ax.set_ylim(-W/1.91, W/1.91)

ax.set_xticks(np.linspace(-3, 3, 13))
ax.set_yticks(np.linspace(-1, 1, 5))
ax.grid()

x = np.linspace(-W, W, 300)
y = np.sin(x)
ax.plot(x,y)

x = np.array([-W, W])
for x0 in np.linspace(-np.pi, np.pi, 100):
  y0 = np.sin(x0); a = np.cos(x0)
  y = a*(x-x0)+y0
  ax.plot(x,y, linewidth=0.5, color='#ff7f0e')

plt.savefig("sin.png",format="png",dpi=300)
plt.show()
```

```python
plt.rcParams['font.size'] = 14 # デフォルトの文字サイズ
r = 5.0
fig, ax = plt.subplots(figsize=(1.91*r,r))
eps = 0.1
W = np.pi # x軸の表示範囲
ax.set_xlim(-W,W)
ax.set_ylim(-W/1.91, W/1.91)

ax.set_xticks(np.linspace(-3, 3, 13))
ax.set_yticks(np.linspace(-1, 1, 5))
ax.grid()

x = np.linspace(-W, W, 300)
y = np.cos(x)
ax.plot(x,y)

x = np.array([-W, W])
for x0 in np.linspace(-np.pi, np.pi, 100):
  y0 = np.cos(x0); a = -np.sin(x0)
  y = a*(x-x0)+y0
  ax.plot(x,y, linewidth=0.5, color='#ff7f0e')

plt.savefig("cos.png",format="png",dpi=300)
plt.show()
```
