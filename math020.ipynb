```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 亀の向きを変える
def L(l):
  l[0] = (l[0]-1) % 6
  return

def R(l):
  l[0] = (l[0]+1) % 6
  return

# 亀を前進させる
def F(l):
  direction = l[0]
  if direction == 0:
    l.insert(1,[l[1][0]+1, l[1][1]  ])
  elif direction == 1:
    l.insert(1,[l[1][0]  , l[1][1]+1])
  elif direction == 2:
    l.insert(1,[l[1][0]-1, l[1][1]+1])
  elif direction == 3:
    l.insert(1,[l[1][0]-1, l[1][1]  ])
  elif direction == 4:
    l.insert(1,[l[1][0]  , l[1][1]-1])
  elif direction == 5:
    l.insert(1,[l[1][0]+1, l[1][1]-1])
  else:
    print("F:error")

# 生成された文字列にしたがって亀を移動させる
def move(l,c):
  if c == '-':
    L(l)
  elif c == '+':
    R(l)
  elif c == 'F':
    F(l)
  else:
    print('move:error')
```

```python
# 文字列を生成するための関数
def generator(c):
  if c == 'F':
    retv = 'F+F--F+F'
  else:
    retv = c
  return retv

def gen_str(s, n):
  for i in range(n):
    #print(s)
    tmp = ''
    for c in s:
      tmp = tmp + generator(c)
    s = tmp
  return s

# パスを生成する関数
def make_path(s):
  l = [ 0, [0,0]]
  for c in s:
    move(l,c)
  # リストを配列に変換
  length = len(l)-1
  A = np.zeros((length, 2))
  for i in range(length):
    A[i,0] = l[length-i][0]
    A[i,1] = l[length-i][1]
  s30 = np.sin(np.pi/3); c30 = np.cos(np.pi/3)
  Base = np.matrix([[1.0, 0.0], [c30, s30]])
  return np.matmul(A, Base)
```
